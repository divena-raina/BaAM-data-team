```python
# Import data cleaning libraries
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Chicago_Population_Counts_20250713.csv')
data
```

```python
data.set_index('Record ID', inplace=True)
data
```

```python
data['Geography Type'].unique()
data = data[data['Geography Type'] != 'Citywide']
data.drop(columns=['Geography Type'], inplace=True)
```

```python
data['Year'].value_counts()
data = data[data['Year'] == 2021]
# Keep only the most recent year
```

```python
data['Zip Code'] = data['Geography']
data.drop(columns=['Geography'], inplace=True)

data.info()
```

```python
# Make entire dataframe integer (can't have float population counts)
data = data.astype(int)
data
```

```python
import geopandas as gpd
ca = gpd.read_file("geo_export_e7c5223a-5c75-4ada-aaea-13af1a1048bb.shp")
zc = gpd.read_file("geo_export_efe0f982-9253-4ead-90c6-a53bfdc9a763.shp")

print("ca CRS:", ca.crs)
print("zc CRS:", zc.crs)

zc.set_crs("EPSG:4326", inplace=True)
ca.set_crs("EPSG:4326", inplace=True)

zc = zc.to_crs(ca.crs)

# Option B: Area-overlap join (more rigorous)
overlay = gpd.overlay(zc, ca[['area_numbe', 'community', 'geometry']], how='intersection')
overlay['area'] = overlay.geometry.area

# For each ZIP, pick community area with largest overlap
idx = overlay.groupby('zip')['area'].idxmax()
zip_comm = overlay.loc[idx, ['zip','area_numbe','community']].rename(columns={
    'ZIP':'zip_code',
    'AREA_NUMBER':'area_numbe',
    'COMMUNITY':'community_area'})


zip_comm['zip'] = zip_comm['zip'].astype(int)
zip_comm['community_area_number'] = zip_comm['area_numbe'].astype(int)
zip_comm = zip_comm[['zip', 'community_area_number', 'community']]


data = data.merge(zip_comm, left_on='Zip Code', how='inner', right_on='zip')
data.drop(columns=['zip'], inplace=True)
```

```python
# Saving the cleaned data to a new CSV file
data.to_csv('../cleaned_data/Chicago_Population_Counts_Cleaned.csv', index=True)
```
